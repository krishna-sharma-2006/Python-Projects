## Library System

```python
from datetime import datetime, timedelta
import pickle as pkl
import json
```

```python
class library:
    
    def __init__(self):
        self.books = {}
        self.users = {}
        self.borrowed_books = {}
        
    def add_book(self,book):
        if book.book_id in self.books:
            raise ValueError("Can't")
        self.books[book.book_id] = book
        
    def register_user(self, user):
        if user.user_id in self.users:
            raise ValueError("Can't")
        self.users[user.user_id] = user

    def issue_book(self,book_id,user_id):
        user = self.users.get(user_id)
        bo0k = self.books.get(book_id)
        if len(user.borrowed_books) == 3 or book_id in user.borrowed_books:
            print("Sorry!...")
        else:
            self.books.get(book_id).mark_issued()
            self.users.get(user_id).borrow_book(book_id)
            self.borrowed_books[book_id] = {
                "user_id" : user_id,
                "issue_date" : datetime.now() - timedelta(days=20),
                "expected_return_date" : datetime.now() - timedelta(days=20) + timedelta(days = 14)
            }

    def load_data(self):
        with open("books.pkl","r") as f:
            self.books = pkl.load("books.pkl")
        with open("users.pkl","r") as users:
            self.users = pkl.load("users.pkl")
        with open("b_books.pkl","r") as b_books:
            self.borrowed_books = pkl.load("b_books.pkl")

    def save_data(self):
        with open("books.pkl","ab") as books:
            pkl.dump(self.books,books)
        with open("users.pkl","ab") as users:
            pkl.dump(self.users,users)
        with open("b_books.pkl","ab") as b_books:
            pkl.dump(self.borrowed_books,b_books)
        
            
    def return_book(self, book_id, user_id):
        book = self.books.get(book_id)
        user = self.users.get(user_id)
        exp_date = self.borrowed_books[book_id]["expected_return_date"]
        return_date = datetime.now()
        if exp_date < return_date :
            user.fine = ((return_date - exp_date).days)*2
            

class Book(library):

    def __init__(self,book_id,author,title):
        self.book_id = book_id
        self.author = author
        self.title = title
        self.is_available = 1
        
    def mark_issued(self):
        self.is_available = 0

    def mark_returned(self):
        self.is_available = 1
        self.issue_date = 0
        self.expected_return_date = 0
        


class user(library):

    def __init__(self, user_id, name):
        self.user_id = user_id
        self.name = name
        self.borrow_limit = 3
        self.fine = 0
        self.borrowed_books = []
        
    def borrow_book(self,book_id):
        self.borrowed_books.append(book_id)

    def return_book(self,book_id):
        self.borrowed_books.remove(book_id)

```
